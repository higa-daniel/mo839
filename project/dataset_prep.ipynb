```python
from skimage import io, filters
import os
from skimage import img_as_ubyte

def apply_gaussian_skimage_folder(input_folder, output_folder, sigma=1.0):
    """
    Apply Gaussian filter using scikit-image
    """
    
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif')
    image_files = [f for f in os.listdir(input_folder) 
                   if f.lower().endswith(image_extensions)]
    
    for image_file in image_files:
        img_path = os.path.join(input_folder, image_file)
        
        try:
            # Read image
            img = io.imread(img_path)
            
            # Apply Gaussian filter
            filtered_img = filters.gaussian(img, sigma=sigma, channel_axis=-1 if img.ndim == 3 else None)
            
            # Convert back to uint8 for saving
            filtered_img_uint8 = img_as_ubyte(filtered_img)
            
            # Save image
            output_path = os.path.join(output_folder, image_file)
            io.imsave(output_path, filtered_img_uint8)
            print(f"Processed: {image_file}")
            
        except Exception as e:
            print(f"Error processing {image_file}: {str(e)}")

# Usage
input_folder = "/home/c-specp25-2/ra992741/Desktop/github/mo839/project/DATASET/TEST/NOT_CRIMINAL/GAUSSIAN_FILTER/input"
output_folder = "/home/c-specp25-2/ra992741/Desktop/github/mo839/project/DATASET/TEST/NOT_CRIMINAL/GAUSSIAN_FILTER/output"
apply_gaussian_skimage_folder(input_folder, output_folder, sigma=1.5)
```

```python
from skimage import io, util
import os
import numpy as np

def apply_salt_pepper_folder(input_folder, output_folder, amount=0.05, salt_vs_pepper=0.5):
    """
    Apply salt and pepper noise to all images in a folder
    
    Args:
        input_folder: Path to folder containing input images
        output_folder: Path to folder where noisy images will be saved
        amount: Proportion of image pixels to replace with noise
        salt_vs_pepper: Ratio of salt vs pepper noise (0.5 = equal amounts)
    """
    
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif')
    image_files = [f for f in os.listdir(input_folder) 
                   if f.lower().endswith(image_extensions)]
    
    for image_file in image_files:
        img_path = os.path.join(input_folder, image_file)
        
        try:
            # Read image
            img = io.imread(img_path)
            
            # Apply salt and pepper noise
            noisy_img = util.random_noise(img, mode='s&p', amount=amount, salt_vs_pepper=salt_vs_pepper)
            
            # Convert from float [0,1] to uint8 [0,255]
            noisy_img = (noisy_img * 255).astype(np.uint8)
            
            # Save image
            output_path = os.path.join(output_folder, image_file)
            io.imsave(output_path, noisy_img)
            print(f"Processed: {image_file}")
            
        except Exception as e:
            print(f"Error processing {image_file}: {str(e)}")

# Usage
input_folder = "/home/c-specp25-2/ra992741/Desktop/github/mo839/project/DATASET/TEST/NOT_CRIMINAL/SALT_AND_PEPPER_FILTER/input"
output_folder = "/home/c-specp25-2/ra992741/Desktop/github/mo839/project/DATASET/TEST/NOT_CRIMINAL/SALT_AND_PEPPER_FILTER/output"
apply_salt_pepper_folder(input_folder, output_folder, amount=0.05, salt_vs_pepper=0.5)
```
