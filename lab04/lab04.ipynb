```python
x = np.arange(0.0, 2.0, 0.01)
y = np.sin(2*np.pi*x)
```

```python
#cria uma figura
fig, ax = plt.subplots(figsize=(9, 5)) #ax eixo dos graficos
                                       #fig espa√ßo figura
                                       
#configurar figura
ax.plot(x, y, color='r', linestyle='-', label='Label 1', marker='o', markevery=5)
ax.plot(x, np.cos(y), color='b', linestyle='--', label='Label 2', marker='o',markevery=5)
ax.plot(x, np.tan(y), color='g', linestyle=':', label='Label 3', marker='o',markevery=5)

#insere titulo na figura
ax.set_title('Este √© o t√≠tulo do Gr√°fico')
#insere rotulo do eixo x
ax.set_xlabel('Este √© o eixo X', size=22)
#insere rotulo do eixo y
ax.set_ylabel('Est√© o eixo Y', size=22)
#aplica grid
plt.grid(True, linestyle='--')
#legenda
plt.legend()
#salvar figura
fig.savefig('exemplo_plot_1.png')
#mostrar figura
plt.show(True)
```

```python
# Scatter
fig, ax = plt.subplots()
y       = np.exp(x)

ax.scatter(x, y, color='orange', marker='x');
plt.grid(True, linestyle=':')
```

4. Calcule o total de sobreviventes, idade m√©dia dos passageiros, e o n√∫mero de homens e mulheres a bordo


```python
total_sobreviventes = (df['alive'] == 'yes').sum()
total_sobreviventes
idade_media = df['age'].mean()
idade_media
```

5. Gere uma visualiza√ß√£o para a quantidade de homens e mulheres a bordo


```python
gender_counts = df['sex'].value_counts().reset_index()
palette = {"male": "#1f77b4", "female": "#ff7f0e"}
sns.barplot(data=gender_counts, x='sex', y='count', palette=palette)
plt.title('Total Number of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

6. Gere uma visualiza√ß√£o que resuma o intevalo das idades dos passageiros considerando seu sexo e se o passageiro sobreviveu.

```python
# Converte a coluna 'survived' em categ√≥rica (0 ‚Üí "N√£o", 1 ‚Üí "Sim")
df['survived'] = df['survived'].map({0: 'N√£o', 1: 'Sim'})

# Cria o boxplot com paleta personalizada
sns.boxplot(
    data=df,
    x='sex',
    y='age',
    hue='survived',
    palette={'N√£o': '#e74c3c', 'Sim': '#2ecc71'}  # vermelho e verde
)

# Ajustes do gr√°fico
plt.title('Distribui√ß√£o das Idades por Sexo e Sobreviv√™ncia')
plt.xlabel('Sexo')
plt.ylabel('Idade')
plt.legend(title='Sobreviveu')
plt.show()
```

7. Gere uma visualiza√ß√£o do tipo boxplot que mostre a tarifa de acordo com a classe dos passageiros


```python
sns.boxplot(
    data=df,
    x='class',
    y='fare',
    palette='Set3'
)
# Ajustes do gr√°fico
plt.title('Distribui√ß√£o das Tarifas por Classe de Passageiro')
plt.xlabel('Classe')
plt.ylabel('Tarifa')
plt.show()
```
